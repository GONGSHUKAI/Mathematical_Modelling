```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
df = pd.read_excel('/.xlsx')
df[''] = pd.to_datetime(df[''])
df.set_index('', inplace=True)
df_resampled = df[''].resample('7D').sum()
df_resampled2 = df[''].resample('8D').sum()
result = seasonal_decompose(df_resampled, model='additive', period=3)
model = SARIMAX(df_resampled, order=(1, 1, 0), seasonal_order=(1, 1, 0, 7), seasonal_periods=7)
results = model.fit(disp=False)
forecast = results.get_forecast(steps=7)
forecast_index = pd.date_range(start=df_resampled.index[-1], periods=7, freq='7D')
merged_data = pd.concat([df_resampled, pd.Series(forecast.predicted_mean+10, index=forecast_index)])
plt.figure(figsize=(12, 6))
plt.plot(merged_data, label='',color='orange')
plt.plot(df_resampled, label='',color='blue')
plt.plot(df_resampled2, color='orange')
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
#plt.savefig(' .png', dpi=300)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
df = pd.read_excel('/.xlsx')
import matplotlib.pyplot as plt
df_diff = df.copy()
df_diff[''] = df[''].diff(1).fillna(0)
result = adfuller(df[''])
result2 = adfuller(df_diff[''])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('ADF Statistic:', result2[0])
print('p-value:', result2[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_diff[''], lags=20)
plt.title(' ACF')
#plt.savefig('ACF.png', dpi=300)
plt.show()
```

```python
plot_pacf(df_diff[''], lags=20)
plt.title(' PACF')
#plt.savefig('PACF.png', dpi=300)
plt.show()
```
