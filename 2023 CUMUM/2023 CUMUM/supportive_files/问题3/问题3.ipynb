```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

table = pd.read_excel('双向优化算法——数据集.xlsx')
table
```

```python
table.T[:4]
```

```python
def profit(price,quant,cost):
    if (quant >2.5):
        return (price-cost)*quant
    else:
        return (price-cost)*quant-(2.5-quant)*cost

for i in range(0,48):
    table.loc[i,'利润']=profit(table.loc[i,'7月1日单品价格'],table.loc[i,'7月1日需求量'],table.loc[i,'7月1日批发价'])


table_sorted = table.sort_values(by='利润',ascending=False)

table_sorted
```

```python
table_esp = table_sorted[:33]

table_esp[['单品名称','7月1日需求量']]
```

```python
from scipy.interpolate import make_interp_spline
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
result=[]
result2=[]


for M in range(20, 48):
    Replenishment = table_sorted[:M]
    T = table_sorted[M:]
    total = T.apply(lambda row: row['7月1日需求量'] * row['7月1日单品价格'], axis=1).sum()
    total2 = Replenishment.apply(lambda row: row['利润'], axis=1).sum()
    result.append(total)
    result2.append(total2)

x = np.arange(20, 48)
result = np.array(result)
result2 = np.array(result2)

spline = make_interp_spline(x, result)
spline2 = make_interp_spline(x, result2)
x_smooth = np.linspace(x.min(), x.max(), 100)
data_smooth = spline(x_smooth)
data_smooth2 = spline2(x_smooth)

graph1 = plt.plot(x_smooth, data_smooth, color='b',label='消费者不满程度S')
graph2 = plt.plot(x_smooth, data_smooth2, color='r',label='超商总利润水平')

plt.xlabel("可售单品总数M")

plt.ylim(0,800)
plt.axvspan(27, 33, facecolor='lightblue', alpha=0.5)
plt.legend(loc = 'upper right')
plt.title('双向优化算法：商超利润与市场需求满足')
#plt.savefig('双向优化.png', dpi=300)
plt.show()

```
